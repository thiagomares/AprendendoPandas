```python
# Analise exploratoria de dados

df.groupby('Tipo').mean(numeric_only=True) # por conta do pandas2, nos temos que passar um parametro falando que o queremos ter como média é os valores numéricos
```

```python
df_preco_tipo = df.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
df_preco_tipo.plot(kind='barh', figsize=(14, 10), color='red')
```

```python
# Removendo imoveis Comerciais
```

```python
df.Tipo.unique() # o Método unique retorna os valores únicos de uma coluna do pandas
```

```python
imoveis_comerciais = ['Conjunto Comercial/Sala', 
                      'Prédio Inteiro', 'Loja/Salão', 
                      'Galpão/Depósito/Armazém', 
                      'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial',
                      'Box/Garagem', 'Chácara',
                      'Loteamento/Condomínio', 'Sítio',
                      'Pousada/Chalé', 'Hotel', 'Indústria']
                                            


dados_imoveis_comerciais = df.query('@imoveis_comerciais in Tipo') # este método retorna apenas os dados que precisamos, assim como fazemos com uma query do sql
dados_imoveis_comerciais
```

```python
dados_imoveis_residenciais = df.query('@imoveis_comerciais not in Tipo')
dados_imoveis_residenciais
```

```python
df_imoveis_residenciais = dados_imoveis_residenciais.groupby('Tipo')[['Valor']].mean().sort_values('Valor')

df_imoveis_residenciais.plot(kind='barh', figsize=(14, 10), color='red')
```

```python
dados_imoveis_residenciais.Tipo.unique()
```

```python
# Agora precisamos contar o numero de tipos que temos

dados_imoveis_residenciais.Tipo.value_counts(normalize=True) # O value_counts() retorna o numero de incidencias daquele valor aparece dentro do dataframe, com isso, temos uma contagem. Para retornar uma porcentagem (fração), basta que inserimos um parametro chamado normalize
```

```python
dados_imoveis_residenciais.Tipo.value_counts(normalize=True).to_frame().sort_values('Tipo')
```

```python
df_percentual_tipo = dados_imoveis_residenciais.Tipo.value_counts(normalize=True).to_frame().sort_values('Tipo')

df_percentual_tipo.plot(kind='bar', figsize=(14, 10), color='red', xlabel='Tipos', ylabel='Percentual', legend='Percentual')
```
